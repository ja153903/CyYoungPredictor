```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./CyYoungEligiblePitchers.csv')
```

```python
# df[df['win'] == 2] # this is the outlier as he was in two leagues
df.loc[(df['player_id'] == 'sutclri01') & (df['year'] == 1984)]['win'] = 1
```

```python
df = df.drop(['player_id', 'year'], axis=1)
```

```python
df['win'] = df['win'].astype(int)
```

```python
df
```

```python
df.columns
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred, y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
print('We have {} % accuracy'.format(cm.diagonal().sum() / cm.sum() * 100))
```
